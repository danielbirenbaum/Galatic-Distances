```python
# Autor: Daniel Birenbaum
# Github: github.com/danielbirenbaum

#Tratando os dados e mostrando a matriz D de distâncias

from sympy import *
from tabulate import tabulate
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
init_printing()

data = open(file="data/data.csv",mode="r",encoding="utf-8")
distances = []
for i in data:
    i = i.split(',')
    distances.append(i)

for i in distances:
    for j in range(len(i)):
        i[j] = int(i[j])

print("Eis a matriz D de distâncias:")
print("")
D = Matrix(distances)
display(D)

#Computando a matriz D normalizada
D /= D.norm().evalf()

```

```python
#Definindo a matriz D2, a matriz com as distâncias quadradas

D2 = D.applyfunc(lambda x : x**2)
display(D2)
```

```python
#Definindo o operador linear P
P = eye(27)
TEMP = (ones(27,27))/27
P -= TEMP
print("Matriz P:")
print("")
display(P)
print("")
eigva = P.eigenvals()
eigvc = P.eigenvects()

print("{Autovalores : Multiplicidade algébrica}:")
print("")
display(eigva)
print("")
print("(Autovalores, multiplicidade algébrica, autovetor(es))")
print("")
display(eigvc)

P = P.evalf()
```

```python
#Definindo a matriz A que induz a matriz P
A = ones(27,27)/27
print("Matriz que induz P:")
display(A)
print("")

# Seja V a matriz de autovetores e L a matriz de autovalores tal que A = VLV^-1
V,L = A.diagonalize()

print("V:")
display(V)
print("")

print("L:")
display(L)
print("")
#Mostra que a diagonalização é válida
print(A == V*L*V.inv())
```

```python
# Cálculo de X_hat

G_hat = -(P*D2*P)/2
G_hat = G_hat.evalf()
print("Matriz G induzida por -2(PGP):")
print("")
display(G_hat)
print("")

eigvals_dict = G_hat.eigenvals()
eigvals = sorted(eigvals_dict.keys(), reverse=True)

print("Autovalores:")
display(eigvals)
print("")

eva1 = eigvals[0]
eva2 = eigvals[1]

eigvects = G_hat.eigenvects()

vec1 = None
vec2 = None

for val, mult, vects in eigvects:
    if val == eva1:
        vec1 = vects[0]
    if val == eva2:
        vec2 = vects[0]
print("\nAutovetor associado ao maior autovalor:")
print("")
display(vec1)
print("\nAutovetor associado ao segundo maior autovalor:")
print("")
display(vec1)

Lambda = diag(eva1, eva2)
Lambda_sqrt = Lambda.applyfunc(lambda x: sqrt(abs(x)))

U = Matrix.hstack(vec1, vec2)

theta_old = -np.pi / 1.4
theta = theta_old + np.pi/2
c, s = np.cos(theta), np.sin(theta)

R = Matrix([
    [c, -s],
    [s,  c]
])


X_sym = R * Lambda_sqrt * U.T
X = np.array(X_sym.tolist(), dtype=float)
img = Image.open("./data/mapaBrasil2.jpg")

plt.figure(figsize=(8, 10))
plt.imshow(img)
plt.axis()


xmin, xmax = X[0,:].min(), X[0,:].max()
ymin, ymax = X[1,:].min(), X[1,:].max()

Xn = X.copy()
Xn[0,:] = (X[0,:] - xmin) / (xmax - xmin)
Xn[1,:] = (X[1,:] - ymin) / (ymax - ymin)

scale_factor = 0.9
target_size = 600 * scale_factor

Xs = Xn[0, :] * target_size + 260
Ys = (1 - Xn[1, :]) * target_size + 110

plt.scatter(Xs, Ys, color='blue', s=50)
plt.axis()
plt.show()
```
